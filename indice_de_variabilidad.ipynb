```python
import pandas as pd
import numpy as np

def indices_variabilidad_multi(df, grupo, variables):
    
    resultados = {}

    for var in variables:
        res = df.groupby(grupo)[var].agg(
            Std_dev=lambda x: x.std(ddof=1),
            Variance='var',
            
            
        )
        resultados[var] = res
    
    # Combina los resultados en un solo dataframe multinivel
    tabla = pd.concat(resultados, axis=1)
    return tabla
```

```python
#df = pd.read_csv('data/raw_coordinates.txt', sep='\t')
df = pd.read_csv('data/X_test_df.csv', sep=',')
vars_raw = ['RawCoord1', 'RawCoord2', 'RawCoord3', 'RawCoord4',
       'RawCoord5', 'RawCoord6', 'RawCoord7', 'RawCoord8', 'RawCoord9',
       'RawCoord10', 'RawCoord11', 'RawCoord12', 'RawCoord13', 'RawCoord14',
       'RawCoord15', 'RawCoord16', 'RawCoord17', 'RawCoord18', 'RawCoord19',
       'RawCoord20', 'RawCoord21', 'RawCoord22', 'RawCoord23', 'RawCoord24',
       'RawCoord25', 'RawCoord26']
df.head()
```

```python
tabla = indices_variabilidad_multi(df, 'Cultivar', vars_raw)
tabla
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np


# --- Estandarizar ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[vars_raw])

# --- PCA (podés cambiar n_components si querés) ---
pca = PCA(n_components=2)
pcs = pca.fit_transform(X_scaled)

# Crear columnas PC1...PC5 en el DataFrame original
for i in range(pcs.shape[1]):
    df[f'PC{i+1}'] = pcs[:, i]

print(df[['Cultivar', 'PC1','PC2']].head())
```

```python
df_res =  indices_variabilidad_multi(df, 'Cultivar', ['PC1','PC2'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cols = [('PC1', 'Std_dev'), ('PC1', 'Variance'),('PC2', 'Std_dev'), ('PC2', 'Variance')]

plt.figure(figsize=(8,6))
sns.heatmap(df_res[cols], annot=True, fmt=".2f", cmap="magma")
plt.title("Variabilidad – PC1 - PC2")
plt.tight_layout()
plt.show()
```
